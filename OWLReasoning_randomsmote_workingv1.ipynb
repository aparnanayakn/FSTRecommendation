```python
def convert_and_round(value):
    try:
        return float(value)
    except (ValueError, TypeError):
        return value  

df = pd.read_csv("test_dataset.csv")
df.fillna(0,inplace=True)
for col in df.columns:
    df[col] = df[col].apply(convert_and_round)

pattern = re.compile(r'[^a-zA-Z0-9]') #removing non-alpha numeric characters from file name

fileObjects = {}
df.shape
```

```python
def add_featurevalue(fileObject, feature_class, mean_name, value):
    feature = feature_class(mean_name + "_feature", namespace=onto)
    mean = mtl.Mean(mean_name + "_mean", namespace=onto)
    feature.qlocation.append(mean)
    mean.hasDatavalue  = [value]
    fileObject.hasQuality.append(feature)
```
