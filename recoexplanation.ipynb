```python
import numpy as np
from sklearn.utils import shuffle

def unify_features_with_explanation(test_sample, train_data, train_labels):
    # Extract features from the test sample
    test_features = test_sample.values
    
    # Find matching indices in the training set
    matching_indices = np.where(np.all(train_data.values == test_features, axis=1))[0]
    
    if matching_indices.any():
        matching_labels = train_labels.iloc[matching_indices]
        matching_instances = train_data.iloc[matching_indices, :]
        
        # Shuffle matching instances to get a representative instance
        matching_instances = shuffle(matching_instances, random_state=42)
        
        # Calculate support and confidence
        total_samples = len(train_data)
        support = len(matching_instances) / total_samples
        confidence = len(matching_instances) / len(train_labels)
        
        # Include a representative matching instance, support, and confidence for explainability
        unified_info = {
            "unified_labels": np.unique(matching_labels),
            "matching_instance": matching_instances.iloc[0, :],  # Get a representative instance
            "support": support,
            "confidence": confidence
        }
        return unified_info
    return {"unified_labels": ["unknown"], "matching_instance": None, "support": 0.0, "confidence": 0.0}
```

```python
X= df_std.iloc[:,:-2]
y = df_std.iloc[:,-1]

label_encoder = LabelEncoder()

for col in X:
    X[col] = label_encoder.fit_transform(X[col])

label_enc = LabelEncoder()
y_encoded = label_enc.fit_transform(y).reshape(-1, 1)

smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=6)
   
predicted_labels = []
for i in range(X_test.shape[0]):
    label = unify_features_with_explanation(X_test.iloc[i], X_train, y_train)
    predicted_labels.append(label['support'])
# Test the function with the first row of the test set
#result = unify_features_with_explanation(X_test.iloc[15], X_train, y_train)
#print(result)

#total_lists_with_unknown = sum("unknown" in sublist for sublist in predicted_labels)
#print(1 - total_lists_with_unknown / len(y_test))

sum1 = 0
for num in predicted_labels:
    sum1 += num 

print("Average support",sum1/(len(predicted_labels)))
```
